```python
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv('../../data/basketball/raw/iowa-state_2025_gamelogs.csv')

print("üèÄ Quick Data Overview")
print("=" * 60)
print(f"Shape: {df.shape}")
print(f"\nColumns ({len(df.columns)}):")

# Let's organize columns by position to understand the structure
for i, col in enumerate(df.columns):
    print(f"{i:2d}. '{col}'")
```

```python
def clean_game_logs(df):
    """
    Clean and structure game log data for modeling.
    
    Returns a DataFrame with:
    - Clear team vs opponent stats
    - Win/loss binary variable
    - Derived features
    - Proper data types
    """
    
    # Create a clean copy
    clean_df = df.copy()
    
    # 1. Parse basic game info
    clean_df['date'] = pd.to_datetime(clean_df['Date'])
    clean_df['opponent'] = clean_df['Opp']
    clean_df['won'] = (clean_df['Rslt'] == 'W').astype(int)
    
    # 2. Parse scores
    clean_df['team_score'] = pd.to_numeric(clean_df['Tm'], errors='coerce')
    clean_df['opp_score'] = pd.to_numeric(clean_df['Opp.1'], errors='coerce')
    clean_df['point_diff'] = clean_df['team_score'] - clean_df['opp_score']
    
    # 3. Parse home/away from the opponent name
    # '@' means away game, 'N' means neutral site
    clean_df['is_home'] = (~clean_df['Opp'].str.contains('@|N', na=False)).astype(int)
    clean_df['is_away'] = clean_df['Opp'].str.contains('@', na=False).astype(int)
    clean_df['is_neutral'] = clean_df['Opp'].str.contains('N', na=False).astype(int)
    
    # Clean opponent name (remove @ and N markers)
    clean_df['opponent'] = clean_df['opponent'].str.replace('@', '').str.replace('N', '').str.strip()
    
    # 4. Rename team stat columns (remove .1 from opponent stats)
    team_stat_cols = [
        'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 
        'eFG%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 
        'AST', 'STL', 'BLK', 'TOV', 'PF'
    ]
    
    opp_stat_cols = [col + '.1' for col in team_stat_cols]
    
    # Rename for clarity
    team_rename = {col: f'team_{col.lower()}' for col in team_stat_cols}
    opp_rename = {col: f'opp_{col.lower().replace(".1", "")}' for col in opp_stat_cols}
    
    clean_df = clean_df.rename(columns={**team_rename, **opp_rename})
    
    # 5. Convert stat columns to numeric
    stat_columns = [f'team_{col.lower()}' for col in team_stat_cols] + \
                   [f'opp_{col.lower()}' for col in team_stat_cols]
    
    for col in stat_columns:
        if col in clean_df.columns:
            clean_df[col] = pd.to_numeric(clean_df[col], errors='coerce')
    
    # 6. Select final columns in logical order
    final_cols = [
        # Game metadata
        'date', 'opponent', 'is_home', 'is_away', 'is_neutral',
        'won', 'team_score', 'opp_score', 'point_diff',
        
        # Team stats
        'team_fg', 'team_fga', 'team_fg%', 
        'team_3p', 'team_3pa', 'team_3p%',
        'team_2p', 'team_2pa', 'team_2p%',
        'team_efg%', 'team_ft', 'team_fta', 'team_ft%',
        'team_orb', 'team_drb', 'team_trb',
        'team_ast', 'team_stl', 'team_blk', 'team_tov', 'team_pf',
        
        # Opponent stats
        'opp_fg', 'opp_fga', 'opp_fg%',
        'opp_3p', 'opp_3pa', 'opp_3p%',
        'opp_2p', 'opp_2pa', 'opp_2p%',
        'opp_efg%', 'opp_ft', 'opp_fta', 'opp_ft%',
        'opp_orb', 'opp_drb', 'opp_trb',
        'opp_ast', 'opp_stl', 'opp_blk', 'opp_tov', 'opp_pf',
        
        # Metadata
        'school_slug', 'season'
    ]
    
    # Keep only columns that exist
    final_cols = [col for col in final_cols if col in clean_df.columns]
    clean_df = clean_df[final_cols]
    
    # 7. Sort by date (important for temporal modeling!)
    clean_df = clean_df.sort_values('date').reset_index(drop=True)
    
    return clean_df

# Clean the data
clean_df = clean_game_logs(df)

print("‚úÖ Data Cleaned!")
print("=" * 60)
print(f"Shape: {clean_df.shape}")
print(f"\nFirst few games:")
clean_df[['date', 'opponent', 'is_home', 'won', 'team_score', 'opp_score', 'point_diff']].head(10)
```

```python
# Save cleaned data
clean_df.to_csv('../../data/basketball/processed/iowa-state_2025_clean.csv', index=False)
print("üíæ Saved clean data to: data/basketball/processed/iowa-state_2025_clean.csv")
```

```python
def create_temporal_sequences(clean_df, window_size=5):
    """
    Create sequences of the last N games for LSTM input.
    
    For each game, we want the stats from the previous window_size games
    to capture momentum, trends, and patterns.
    
    Parameters:
    -----------
    clean_df : DataFrame
        Cleaned game log data
    window_size : int
        Number of previous games to include (default: 5)
        
    Returns:
    --------
    sequences : numpy array
        Shape: (num_games, window_size, num_features)
    targets : numpy array
        Shape: (num_games,) - win/loss for each game
    game_indices : list
        Which games have valid sequences (first window_size games don't)
    """
    
    # Select features for LSTM (only numerical stats, not metadata)
    feature_cols = [col for col in clean_df.columns if col.startswith('team_') or col.startswith('opp_')]
    feature_cols += ['point_diff', 'is_home', 'is_away', 'is_neutral']
    
    print(f"üìä Using {len(feature_cols)} features for LSTM")
    
    # Extract feature matrix
    X = clean_df[feature_cols].values
    y = clean_df['won'].values
    
    # Create sequences
    sequences = []
    targets = []
    game_indices = []
    
    for i in range(window_size, len(X)):
        # Get the last window_size games BEFORE this game
        sequence = X[i-window_size:i]
        target = y[i]
        
        sequences.append(sequence)
        targets.append(target)
        game_indices.append(i)
    
    sequences = np.array(sequences)
    targets = np.array(targets)
    
    print(f"\n‚úÖ Created {len(sequences)} sequences")
    print(f"   Shape: {sequences.shape}")
    print(f"   (num_sequences, window_size, num_features)")
    
    return sequences, targets, game_indices, feature_cols

# Create sequences with last 5 games
sequences_5, targets_5, indices_5, feature_names = create_temporal_sequences(clean_df, window_size=5)

print(f"\nüèÄ Example: For game #{indices_5[0]}, we use games #{indices_5[0]-5} through #{indices_5[0]-1}")
print(f"   Target: {'WIN' if targets_5[0] == 1 else 'LOSS'}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split

# Check CUDA
print(f"üöÄ CUDA Available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"   Device: {torch.cuda.get_device_name(0)}")

# Convert to PyTorch tensors
X_tensor = torch.FloatTensor(sequences_5)
y_tensor = torch.FloatTensor(targets_5)

# Split into train/test (80/20)
X_train, X_test, y_train, y_test = train_test_split(
    X_tensor, y_tensor, test_size=0.2, random_state=42, shuffle=True
)

print(f"\nüìä Data Split:")
print(f"   Training: {X_train.shape[0]} games")
print(f"   Testing: {X_test.shape[0]} games")

# Define the LSTM model
class BasketballLSTM(nn.Module):
    def __init__(self, input_size, hidden_size=64, num_layers=2, dropout=0.2):
        """
        LSTM for basketball game outcome prediction.
        
        Architecture:
        - LSTM layers to capture temporal patterns
        - Dropout for regularization
        - Fully connected layers for final prediction
        
        Parameters:
        -----------
        input_size : int
            Number of features per game (48 in our case)
        hidden_size : int
            Size of LSTM hidden state
        num_layers : int
            Number of stacked LSTM layers
        dropout : float
            Dropout probability
        """
        super(BasketballLSTM, self).__init__()
        
        # LSTM layers
        self.lstm = nn.LSTM(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0
        )
        
        # Fully connected layers
        self.fc1 = nn.Linear(hidden_size, 32)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(dropout)
        self.fc2 = nn.Linear(32, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        # x shape: (batch_size, sequence_length, input_size)
        # LSTM output
        lstm_out, (hidden, cell) = self.lstm(x)
        
        # Take the last timestep's output
        last_output = lstm_out[:, -1, :]  # Shape: (batch_size, hidden_size)
        
        # Fully connected layers
        out = self.fc1(last_output)
        out = self.relu(out)
        out = self.dropout(out)
        out = self.fc2(out)
        out = self.sigmoid(out)  # Output probability between 0 and 1
        
        return out

# Initialize model
input_size = sequences_5.shape[2]  # 48 features
model = BasketballLSTM(input_size=input_size, hidden_size=64, num_layers=2)

# Move to GPU!
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

print(f"\nüèÄ Basketball LSTM Model")
print("=" * 60)
print(model)
print(f"\nüìä Total parameters: {sum(p.numel() for p in model.parameters()):,}")
print(f"üî• Training on: {device}")
```

```python
# Training setup
criterion = nn.BCELoss()  # Binary Cross Entropy for win/loss
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Move data to GPU
X_train = X_train.to(device)
y_train = y_train.to(device)
X_test = X_test.to(device)
y_test = y_test.to(device)

# Training loop
num_epochs = 100
batch_size = 4

print("\nüèÄ Training Basketball LSTM")
print("=" * 60)

train_losses = []
test_losses = []

for epoch in range(num_epochs):
    model.train()
    
    # Forward pass
    outputs = model(X_train).squeeze()
    loss = criterion(outputs, y_train)
    
    # Backward pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    # Evaluation
    model.eval()
    with torch.no_grad():
        test_outputs = model(X_test).squeeze()
        test_loss = criterion(test_outputs, y_test)
        
        # Calculate accuracy
        train_preds = (outputs > 0.5).float()
        test_preds = (test_outputs > 0.5).float()
        
        train_acc = (train_preds == y_train).float().mean()
        test_acc = (test_preds == y_test).float().mean()
    
    train_losses.append(loss.item())
    test_losses.append(test_loss.item())
    
    # Print progress every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}] "
              f"Train Loss: {loss.item():.4f} | Train Acc: {train_acc:.2%} | "
              f"Test Loss: {test_loss.item():.4f} | Test Acc: {test_acc:.2%}")

print("\n‚úÖ Training Complete!")

# Final evaluation
model.eval()
with torch.no_grad():
    final_outputs = model(X_test).squeeze()
    final_preds = (final_outputs > 0.5).float()
    final_acc = (final_preds == y_test).float().mean()
    
    print(f"\nüéØ Final Test Accuracy: {final_acc:.2%}")
    print(f"\nPredictions vs Actual:")
    for i in range(len(y_test)):
        actual = "WIN" if y_test[i].item() == 1 else "LOSS"
        predicted = "WIN" if final_preds[i].item() == 1 else "LOSS"
        confidence = final_outputs[i].item()
        print(f"   Game {i+1}: Predicted {predicted} ({confidence:.2%}) | Actual: {actual}")
```

```python
# Extract LSTM hidden states as features
def extract_temporal_features(model, X):
    """
    Extract LSTM hidden state features from sequences.
    These capture the learned temporal patterns.
    """
    model.eval()
    with torch.no_grad():
        # Run through LSTM
        lstm_out, (hidden, cell) = model.lstm(X)
        
        # Use the last hidden state as the temporal feature
        # Shape: (batch_size, hidden_size)
        temporal_features = lstm_out[:, -1, :].cpu().numpy()
    
    return temporal_features

# Extract features for all sequences
all_sequences = torch.FloatTensor(sequences_5).to(device)
temporal_features = extract_temporal_features(model, all_sequences)

print(f"‚úÖ Extracted temporal features!")
print(f"   Shape: {temporal_features.shape}")
print(f"   (num_games, temporal_feature_dim)")

# Add these to our clean dataframe
# Remember: first 5 games don't have temporal features
clean_df_with_temporal = clean_df.iloc[5:].copy().reset_index(drop=True)

# Add each temporal feature as a column
for i in range(temporal_features.shape[1]):
    clean_df_with_temporal[f'lstm_feature_{i}'] = temporal_features[:, i]

print(f"\nüìä Enhanced DataFrame shape: {clean_df_with_temporal.shape}")
print(f"   Original features + {temporal_features.shape[1]} LSTM features")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Prepare data - exclude metadata columns
feature_cols = [col for col in clean_df_with_temporal.columns 
                if col not in ['date', 'opponent', 'school_slug', 'season', 'won']]

X_all = clean_df_with_temporal[feature_cols].values
y_all = clean_df_with_temporal['won'].values

# Split
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(
    X_all, y_all, test_size=0.2, random_state=42
)

print("üå≤ Training Random Forest with LSTM Features")
print("=" * 60)

# Model 1: Using ALL features (stats + LSTM)
rf_with_lstm = RandomForestClassifier(n_estimators=100, random_state=42, max_depth=5)
rf_with_lstm.fit(X_train_rf, y_train_rf)
pred_with_lstm = rf_with_lstm.predict(X_test_rf)
acc_with_lstm = accuracy_score(y_test_rf, pred_with_lstm)

# Model 2: Using ONLY raw stats (no LSTM features)
raw_feature_cols = [col for col in feature_cols if not col.startswith('lstm_feature_')]
X_train_raw = clean_df_with_temporal.iloc[X_train_rf.shape[0]:][raw_feature_cols].values[:X_train_rf.shape[0]]
X_test_raw = clean_df_with_temporal.iloc[-X_test_rf.shape[0]:][raw_feature_cols].values

rf_without_lstm = RandomForestClassifier(n_estimators=100, random_state=42, max_depth=5)
rf_without_lstm.fit(X_train_rf[:, :len(raw_feature_cols)], y_train_rf)
pred_without_lstm = rf_without_lstm.predict(X_test_rf[:, :len(raw_feature_cols)])
acc_without_lstm = accuracy_score(y_test_rf, pred_without_lstm)

print(f"\nüìä Results:")
print(f"   Accuracy WITHOUT LSTM features: {acc_without_lstm:.1%}")
print(f"   Accuracy WITH LSTM features:    {acc_with_lstm:.1%}")
print(f"   Improvement: {(acc_with_lstm - acc_without_lstm):.1%}")

print(f"\nüéØ Predictions with LSTM features:")
print(classification_report(y_test_rf, pred_with_lstm, target_names=['LOSS', 'WIN']))
```
